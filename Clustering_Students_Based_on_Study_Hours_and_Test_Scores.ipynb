```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Clustering Students Based on Study Hours and Test Scores
# Feature 1 (Study Hours): How many hours each student studies per week.
# Feature 2 (Test Scores): The score each student achieves on an exam.

# Sample Data:
# Study Hours (Feature 1)
# Test Scores (Feature 2)
# 5   -> 55
# 8   -> 65
# 2   -> 40
# 10  -> 85
# 7   -> 70

# Descriptions of the 4 clusters and their behavioral patterns:
# Cluster 1 (High Performers - High Achievers): High study hours, high test scores, disciplined, consistent, and focused students.
# Cluster 2 (Moderate Performers - Consistent Learners): Moderate study hours, decent test scores, balanced, consistent, and involved in other activities.
# Cluster 3 (Low Performers - Underperformers): Low study hours, low test scores, need improvement in study habits, struggle with consistency.
# Cluster 4 (Inconsistent Performers - Inconsistent Learners): Variable study hours and test scores, inconsistent results, require better study strategies.

# Example data for students
study_hours = np.array([5, 8, 2, 10, 7, 6, 3, 9, 4, 7]).reshape(-1, 1)  # Feature 1: Study Hours
test_scores = np.array([55, 65, 40, 85, 70, 60, 45, 80, 50, 75]).reshape(-1, 1)  # Feature 2: Test Scores

# Combine features
X = np.hstack([study_hours, test_scores])

# K-Means: The algorithm groups students into 4 clusters based on similarities in study hours and test scores.
kmeans = KMeans(n_clusters=4)  # We now specify 4 clusters
kmeans.fit(X)
clusters = kmeans.predict(X)

# Plot the clusters
plt.scatter(study_hours, test_scores, c=clusters, cmap='viridis', marker='o')
plt.xlabel('Study Hours')
plt.ylabel('Test Scores')
plt.title('Student Clusters Based on Study Hours and Test Scores')
plt.show()
```
