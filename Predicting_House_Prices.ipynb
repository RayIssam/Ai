```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Scenario: Predicting House Prices
# We want to predict the price of houses (the label) based on two features.

# Feature 1: Size of the house (in square feet)
# Feature 2: Number of bedrooms

# Example Data:
# Feature 1 (Size in square feet): This represents the size of each house.
# Feature 2 (Number of bedrooms): This is the number of bedrooms in each house.
# Label (House price): The actual price of the house in thousands of dollars.

# Sample Data (for illustration):
# Size (sq ft)  |  Number of Bedrooms  |  Price ($1000)
# 1500          |  3                   |  400
# 1800          |  4                   |  450
# 1200          |  2                   |  350
# 2000          |  4                   |  500
# 1700          |  3                   |  420

# How the data looks:
# Feature 1 (Size): 1500, 1800, 1200, 2000, 1700
# Feature 2 (Bedrooms): 3, 4, 2, 4, 3
# Label (Price): 400, 450, 350, 500, 420

# Python Example Using this Real-life Data:
# Here’s how we could set up and train a regression model using this data:

# Real-life data for the house prices problem
size = np.array([1500, 1800, 1200, 2000, 1700]).reshape(-1, 1)  # Feature 1: Size in square feet
bedrooms = np.array([3, 4, 2, 4, 3]).reshape(-1, 1)  # Feature 2: Number of bedrooms
price = np.array([400, 450, 350, 500, 420]).reshape(-1, 1)  # Label: Price in thousands of dollars

# Combine features into one matrix
X = np.hstack([size, bedrooms])

# Train the regression model
model = LinearRegression()  # Instantiate the Linear Regression model
model.fit(X, price)  # Fit the model with the features (X) and the labels (price)

# Make predictions
price_pred = model.predict(X)  # Predict house prices based on size and bedrooms

# Visualize the results in 3D space
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the real house prices (in blue)
ax.scatter(size, bedrooms, price, c='b', marker='o', label='Real Prices')

# Plot the predicted house prices (in red)
ax.scatter(size, bedrooms, price_pred, c='r', marker='^', label='Predicted Prices')

# Label the axes for better understanding
ax.set_xlabel('Size (sq ft)')  # X-axis: Size of the house
ax.set_ylabel('Number of Bedrooms')  # Y-axis: Number of bedrooms
ax.set_zlabel('Price ($1000)')  # Z-axis: Price in thousands of dollars

# Display the legend to differentiate between real and predicted prices
plt.legend()

# Show the 3D plot
plt.show()

# Explanation:
# Feature 1 (Size) and Feature 2 (Number of Bedrooms) are combined to predict house prices.
# We fit the linear regression model and plot both the actual and predicted house prices in 3D.
# This is a simplified example of how you can use two features to predict a label (house price).
# You can expand this by adding more data or more features (like location, age of the house, etc.).
```
