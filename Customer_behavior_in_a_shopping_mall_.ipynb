```python
'''
Feature 1: Time spent in clothing stores (randomized).
Feature 2: Time spent in the food court (a sine wave to create the spiral effect).
Latent Variable: Likelihood to buy, going from 0 to 1.

Each point on the plot is a customer.
a point might show that one customer spent 30 minutes in clothing and 10 minutes in the food court.
Darker points might mean they are less likely to buy.
Lighter points might mean they are more likely to buy.
The shape (spiral or circle) shows us that there are patterns in the way people behave. Maybe customers who spend more time in clothing are more likely to buy something, or maybe there's a trend of people starting in one area and moving to another.
'''

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Generating fake data
np.random.seed(42)
n_points = 200
feature_1 = np.linspace(0, 50, n_points) + np.random.normal(0, 5, n_points)
feature_2 = 30 * np.sin(feature_1 / 10) + np.random.normal(0, 3, n_points)
latent_variable = np.linspace(0, 1, n_points)

# Creating a DataFrame
data = pd.DataFrame({
    'Feature 1': feature_1,
    'Feature 2': feature_2,
    'Latent Variable': latent_variable
})

# Plotting
plt.figure(figsize=(8, 6))
scatter = sns.scatterplot(
    x='Feature 1',
    y='Feature 2',
    hue='Latent Variable',
    palette='viridis',
    data=data,
    s=60
)
scatter.set_title('Customer Behavior in a Shopping Mall')
plt.colorbar(scatter.collections[0], label="Likelihood to Buy")
plt.show()
```
